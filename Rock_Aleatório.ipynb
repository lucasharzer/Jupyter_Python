```python
# Escolha das 5 Músicas

import pygame as py
from time import sleep
from datetime import date


def Linha():
    print('\033[1;34m=--------------------------------------=\033[m')
    
def Titulo(msg):
    print(msg.center(40))
    
# Inicio 

Linha()
Titulo('Escolha o melhor Rock!!!♫')
Linha()

# Nome

nome = str(input('Informe seu nome: ')).capitalize()

# Cálculo da idade

ano_nascimento = int (input('Em que ano você nasceu: '))
ano_atual = date.today().year
idade = ano_atual - ano_nascimento

print(f'Seja Bem-Vindo(a) \033[1;32m{nome}\033[m! Você tem \033[1;32m{idade}\033[m anos de idade.')
sleep(2)
print('\033[1;35m\nAguarde ...\n\033[m')
sleep(2)

# Apresentar opções

while True:

    Linha()
    Titulo('Opções: ')
    Linha()
    print('1 -> Primeira Música\n2 -> Segunda Música\n3 -> Terceira Música\n4 -> Quarta Música\n5 -> Quinta Música\n6 -> Sair')
    Linha()

    # Escolha da música

    sleep(3)
    try:
        escolha = int(input('Escolha uma opção: '))

        # Condições

        musica = ''
        name = ''

        if escolha == 1:
            musica = 'musica1.mp3'
            name = 'Five to one (The Doors - 1968)'
        elif escolha == 2:
            musica = 'musica2.mp3'
            name = 'Back in black (AC/DC - 1980)'
        elif escolha == 3:
            musica = 'musica3.mp3'
            name = "You're Heartbreaker (Remastered - 1979)"
        elif escolha == 4:
            musica = 'musica4.mp3'
            name = 'Paranoid (Black Sabbath - 1970)'
        elif escolha == 5:
            musica = 'musica5.mp3'
            name = 'Comfortable Numb (Pink Floid - 1979)'
        elif escolha == 6:
            print('\033[1;35m\nSaindo ...\n\033[m')
            sleep(2)
            break
        else:
            print('\033[1;31m\nEssa opção não existe, opções existentes abaixo.\n\033[m')
    except (ValueError):
        escolha = 0
        print('\033[1;31m\nTivemos problemas com o tipo de valor que você digitou, tente novamente.\n\033[m')
            
    # Rodando a música

    if escolha == 1 or escolha == 2 or escolha == 3 or escolha == 4 or escolha == 5:
        Linha()
        print(f'Você escolheu a opção \033[1;32m{escolha}\033[m\nRodando a música ♪\033[1;32m{name}\033[m♪!!!')
        sleep(2)
        py.mixer.init()
        py.mixer.music.load(musica)
        py.mixer.music.play()

        while py.mixer.music.get_busy():
            while True:
                pause = str(input("\nDigite \033[1;31m'P'\033[m para pausar, \033[1;32m'D'\033[m para despausar ou \033[1;36m'V'\033[m para voltar: ")).strip().upper()
                if pause == 'P':
                    py.mixer.music.pause()
                    Linha()
                    print('\033[1;31mPausado!\033[m')
                elif pause == 'D':
                    py.mixer.music.unpause()
                    Linha()
                    print('\033[1;32mDespausado!\033[m')
                elif pause == 'V':
                    py.mixer.music.pause()
                    print('\033[1;35m\nVoltando ...\n\033[m')
                    sleep(2)
                    break
            break
print(f'\033[1;32mObrigado por participar {nome} !!!')
```
